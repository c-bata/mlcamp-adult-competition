```python
import pandas as pd
```

```python
headers = ["age", "workclass", "fnlwgt", "education", "education-num", "marital-status", "occupation", "relationship", "race", "sex", "capital-gain", "capital-loss", "hours-per-week", "native-country", "target"]
adult_data = pd.read_csv('./data/adult.data', header=None, names=headers, index_col=False, skipinitialspace=True)
adult_test = pd.read_csv('./data/adult.test', header=None, names=headers, index_col=False, skipinitialspace=True)
```

## 欠損値をNoneにする

```python
def set_missing_values(data):
    data.loc[data['workclass'] == '?', 'workclass'] = None
    data.loc[data['occupation'] == '?', 'occupation'] = None
    data.loc[data['native-country'] == '?', 'native-country'] = None
```

## トレーニングデータセット

```python
false_case = adult_data['target'][0]
false_case
```

```python
train_x = adult_data.drop(['target'], axis=1)
train_y = adult_data['target'].apply(lambda x: 0 if x == false_case else 1)
```

```python
train_x
```

```python
train_y
```

```python
set_missing_values(train_x)
```

```python
print(train_x['workclass'].unique())
```

```python
print(train_x['occupation'].unique())
```

```python
train_x
```

```python
train_x.to_csv('data/train_x.csv', index=False)
train_y.to_csv('data/train_y.csv', index=False, header=False)
```

## テストデータセット

```python
false_case = adult_test['target'][0]
false_case
```

```python
test_x = adult_test.drop(['target'], axis=1)
test_y = adult_test['target'].apply(lambda x: 0 if x == false_case else 1)
```

```python
set_missing_values(test_x)
```

```python
test_x.to_csv('data/test_x.csv', index=False)
test_y.to_csv('data/test_y.csv', index=False, header=False)
```
